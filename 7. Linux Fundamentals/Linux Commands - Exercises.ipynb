# **Linux Commands - Exercises** 📒

Her er en kort forklaring og et eksempel på bruk av hvert av Linux-kommandoene:

### Installer sudo
```bash
apt-get install sudo
```
**Forklaring:** Installerer sudo-pakken.
**Eksempel:**
```bash
sudo apt-get install sudo
```

### apt-get update
```bash
apt-get update
```
**Forklaring:** Oppdaterer pakkelisten fra konfigurerte repositorier.
**Eksempel:**
```bash
sudo apt-get update
```

### apt-get upgrade
```bash
apt-get upgrade
```
**Forklaring:** Oppgraderer alle installerte pakker til nyeste versjon.
**Eksempel:**
```bash
sudo apt-get upgrade
```

### ls (-l, -a,)
#### Bruk ls for å se rettighetene på de skjulte filene i root
```bash
ls -la /root
```
**Forklaring:** Viser alle filer (inkludert skjulte) og deres detaljer i root-katalogen.

#### Bruk ls for å se inodene til filene i /etc, uten å endre hvilken mappe du er i
```bash
ls -li /etc
```
**Forklaring:** Viser filene i /etc med deres inode-numre.

### cd
#### Beveg deg fra root til hjemmemappen din med én kommando
```bash
cd ~
```
**Forklaring:** Endrer nåværende arbeidskatalog til din hjemmekatalog.

#### Beveg deg ut av hjemmemappen din med cd «..». Hvor er du nå?
```bash
cd ..
pwd
```
**Forklaring:** Beveger deg opp ett nivå i katalogstrukturen og viser nåværende katalog.

### pwd
#### Skriv «su - <BrukerPåSystemetDitt>». Bruk så pwd for å sjekke hvor du er
```bash
su - username
pwd
```
**Forklaring:** Logger inn som en annen bruker og viser arbeidskatalogen for den brukeren.

#### Skriv whoami for å se hvem du er. Skriv bash for å starte bash.
```bash
whoami
bash
```
**Forklaring:** Viser hvem du er logget inn som, og starter Bash shell.

### Clear
```bash
cat /etc/shadow
clear
```
**Forklaring:** Viser innholdet i `/etc/shadow` og bruker `clear` for å rydde terminalskjermen.

### Sudo
#### Logg på med en annen bruker med su, og prøv å installere tree med «apt-get install tree». Legg så til brukeren i sudo mappa med «visudo». Logg på brukeren og prøv å laste ned igjen, med sudo kommandoen.
```bash
su - otheruser
apt-get install tree
sudo visudo  # Legg til "otheruser ALL=(ALL) ALL"
exit
sudo apt-get install tree
```
**Forklaring:** Logger på som en annen bruker, forsøker å installere 'tree' uten sudo, legger til brukeren i sudoers-filen, og forsøker igjen med sudo.

### Cp
#### Kopier hjemmemappen din til / og gi den navnet «HomeCopy». Gjør dette i én kommando
```bash
cp -r ~ /HomeCopy
```
**Forklaring:** Kopierer hjemmekatalogen til rotkatalogen og gir den navnet HomeCopy.

### Mv
#### Flytt HomeCopy tilbake til /home/ og endre navnet til «<Brukernavn>2».
```bash
mv /HomeCopy /home/<username>2
```
**Forklaring:** Flytter og omdøper katalogen HomeCopy til brukerens hjemmekatalog med et nytt navn.

### Mkdir
```bash
mkdir ~/NyMappe
```
**Forklaring:** Oppretter en ny mappe kalt 'NyMappe' i din hjemmekatalog.

### Rmdir
```bash
rmdir ~/NyMappe
```
**Forklaring:** Fjerner mappen 'NyMappe'.

### Tree
```bash
tree ~
```
**Forklaring:** Viser en trestruktur av din hjemmekatalog.

### Cat
```bash
cat /etc/passwd
```
**F

orklaring:** Viser innholdet i filen '/etc/passwd' som inneholder brukerinformasjon.

### ln (-s)
#### Lag en snarvei til hjemmemappen din
```bash
ln -s ~ ~/Desktop/HomeShortcut
```
**Forklaring:** Lager en symbolsk lenke (snarvei) til hjemmekatalogen på skrivebordet.

### Top
```bash
top
```
**Forklaring:** Viser en dynamisk liste over kjørende prosesser.

### grep (-i, -v)
#### Skriv ut alle linjer fra /etc/passwd som har «u» i seg
```bash
grep "u" /etc/passwd
```
**Forklaring:** Viser alle linjer fra '/etc/passwd' som inneholder bokstaven 'u'.

### adduser
```bash
adduser MidlertidigBruker
```
**Forklaring:** Legger til en ny bruker med navnet 'MidlertidigBruker'.

### Deluser
```bash
deluser MidlertidigBruker
```
**Forklaring:** Sletter brukeren 'MidlertidigBruker'.

### Groupadd
```bash
groupadd NyGruppe
```
**Forklaring:** Oppretter en ny gruppe kalt 'NyGruppe'.

### usermod (-a, -G, -g)
```bash
usermod -aG NyGruppe username
```
**Forklaring:** Legger til brukeren 'username' i gruppen 'NyGruppe'.

### Delgroup
```bash
delgroup NyGruppe
```
**Forklaring:** Sletter gruppen 'NyGruppe'.

### chmod (-R)
```bash
chmod -R 755 ~/Documents
```
**Forklaring:** Setter lese-, skrive- og kjøretilganger for eieren, og lese- og kjøretilganger for andre, rekursivt i 'Documents'-mappen.

### Chgrp
```bash
chgrp NyGruppe ~/Documents
```
**Forklaring:** Endrer gruppen for mappen 'Documents' til 'NyGruppe'.

### Chown
```bash
chown username ~/Documents
```
**Forklaring:** Endrer eieren av mappen 'Documents' til 'username'.

### Reboot
```bash
reboot
```
**Forklaring:** Restarter datamaskinen.

### Vi
```bash
vi ~/Fortelling.txt
```
**Forklaring:** Åpner eller oppretter filen 'Fortelling.txt' i 'vi'-editoren.

### Nano
```bash
nano ~/Fortelling2.txt
```
**Forklaring:** Åpner eller oppretter filen 'Fortelling2.txt' i 'nano'-editoren.

### crontab (-e)
```bash
crontab -e
# Legg til: "30 23 * * 5 cp -r ~ /var/"
```
**Forklaring:** Setter opp en crontab-jobb som kopierer hjemmekatalogen til '/var/' hver fredag kl. 23:30.

### tail (-n)
```bash
tail -n 3 /etc/passwd
```
**Forklaring:** Viser de tre siste linjene i filen '/etc/passwd'.

### head (-n)
```bash
head -n 5 /etc/group
```
**Forklaring:** Viser de første fem linjene i filen '/etc/group'.

### ps (-aux)
```bash
ps aux
```
**Forklaring:** Viser en detaljert liste over kjørende prosesser.

### man
```bash
man ls
```
**Forklaring:** Viser manualen (hjelpefilen) for kommandoen 'ls'.

### bg
```bash
some_command &
bg
```
**Forklaring:** Starter en kommando i bakgrunnen.

### Fg
```bash
fg
```
**Forklaring:** Henter en bakgrunnsprosess til forgrunnen.

### Kill
```bash
kill 1234
```
**Forklaring:** Avslutter prosessen med prosess-ID 1234.

Dette gir en oversikt og eksempler for hvert av kommandoene. Hvis du trenger mer informasjon om noen av dem, bare gi beskjed!

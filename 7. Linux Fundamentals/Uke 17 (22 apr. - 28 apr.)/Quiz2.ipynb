#### **1. Hva er en inode?**
* En inode er et datasett som inneholder metadata om filer. 

**Forklaring**
I Linux er en **inode** (indeksnode) en datastruktur som lagrer nesten all informasjon om en fil untatt navnet og den faktiske datainnholdet:
* teksten av filen
* bildedata
